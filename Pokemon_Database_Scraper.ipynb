## Now that we can see all the data we need is correct, lets make a CSV!

### But first I will convert those earlier loops into functions.

```python
def PokemonNames(soup):
    """Returns a list of the names of the Generation 4 Pokemon."""
    
    PokemonNames = soup
    PokemonNamesList = []
    
    for i in range(len(PokemonNames)):
        PokemonNamesList.append(str(PokemonNames[i]).split('>')[1].split('<')[0])
    
    return PokemonNamesList
```

```python
def PokemonRegDexNum(soup):
    """Returns a list of the Regional Pokedex entry numbers of the Pokemon in Generation 4."""
    
    DexNum = soup
    DexNumList = []
    
    for i in range(len(DexNum)):
        DexNumList.append(str(DexNum[i]).split('>')[1].split('<')[0])
    
    return DexNumList
```

```python
def PokemonTyping(soup):
    """Returns a list of the Typings of all of the Generation 4 Pokemon."""
    
    Montyping = soup
    MontypingList = []
    
    for i in range(len(Montyping)):
        if len(str(Montyping[i]).split('>')) < 6:
            MontypingList.append(str(Montyping[i]).split('>')[2].split('<')[0])
        elif len(str(Montyping[i]).split('>')) > 5:
            MontypingList.append(str(Montyping[i]).split('>')[2].split('<')[0] + '/' + str(Montyping[i]).split('>')[4].split('<')[0])

    return MontypingList
```

### Now to create variables to house the result sets from soup, that we will pass to these functions.

```python
RegDexNum = soup.find_all('small')
RegDexNum = RegDexNum[0::2]

Typing = soup.find_all('small')
Typing = Typing[1::2]

names = soup.find_all(class_='ent-name')
```

### Now let's call those functions, and create a DataFrame with the returned lists.

```python
nameslist = PokemonNames(names)
dexnumlist = PokemonRegDexNum(RegDexNum)
typelist = PokemonTyping(Typing)

print(nameslist, dexnumlist, typelist)
```

```python
data = {'Num': dexnumlist,
        'Name': nameslist,
        'Typing': typelist}

nametype_df = pd.DataFrame(data)
```

### Perfect! Now I must scrape the stats of each individual Pokemon and append it column-wise to this DataFrame.

### To start I have to make a function to create a list of URL's that the scraper must pull data from.
### To do that I will be making an altered 'NameList' where I will be cleaning this data by removing spaces and periods from Pokemon names. 

```python
alteredName = []

for i in range(len(nameslist)):
    j = nameslist[i].replace(' ', '-')
    j = j.replace('.','')
    alteredName.append(j)

print(alteredName)
```

```python
def Pokeurls():
    """Takes the name of each Pokemon from the 'nameslist' variable then appends 
       it to the base url then stores the urls as strings in a list."""

    base_url = 'https://pokemondb.net/pokedex/'
    listofurls = []
    
    for i in range(len(alteredName)):
        listofurls.append(str(base_url + alteredName[i]))
        
    return listofurls

```

```python
def PokeData(urls):
    
    listofurls = urls
    stats_list = []
    count = 0

    for i in range(len(listofurls)):
        count = 0
        T_url = listofurls[i]
        T_res = requests.get(T_url)
        T_soup = BeautifulSoup(T_res.text, 'html.parser')
        T_soup = T_soup.find_all(class_='vitals-table')
        T_soup = T_soup[3].find_all(class_='cell-num')[0::3]
        
        hp = list(T_soup[0]) # Looks like we need to make several small lists to appened to a main and return it
        atk = list(T_soup[1])
        dfn = list(T_soup[2])
        spatk = list(T_soup[3])
        spdfn = list(T_soup[4])
        spd = list(T_soup[5])
    
        stat = [list(e) for e in zip(hp, atk, dfn, spatk, spdfn, spd)]
        stats_list.append(stat)

    return stats_list

stat_table = PokeData(Pokeurls())
final_stat_list = []
for i in range(len(stat_table)):
    final_stat_list.append(stat_table[i][0])

final_stat_list
```

```python
stats_df = pd.DataFrame(final_stat_list, columns=['HP', 'ATK', 'DFN', 'SPATK', 'SPDFN', 'SPD'])
```

### Now let us extend these dataframes into one dataframe and export it as a CSV file and get to work.

```python
PokeDF = pd.concat([nametype_df, stats_df], axis=1)
PokeDF
```
